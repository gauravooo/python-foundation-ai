```python
import numpy as np
arr = np.array([1,2,3,4,5])
print(arr)
print(type(arr))
```

```python
print(f"Shape: {arr.shape}")
print("Data Type:",arr.dtype)
print("Size:",arr.size)
```

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
print("Addition:",a+b)
print("Multiplication:",a*b)
print("Dot Product:", np.dot(a,b))
```

```python
matrix = np.array([[1,2,3],[4,5,6]])
print ("Matrix:",matrix.shape)
print("Dimensions:",matrix.ndim)
print(matrix)
```

```python
print("Element at [0,1]:",matrix[0,1])
print("Element row:",matrix[0,:])
print("Element column:",matrix[:,1])
```

```python
print("Matrix *2 :\n",matrix*2)
print("Matrix+10 :\n",matrix + 10)
print(matrix.sum())
print(matrix.sum(axis=1))
print(matrix.sum(axis=0))
```

```python
a = np.array([[1,2,3],[4,5,6]])
b = np.array([[7,8],[10,11],[9,12]])
print(np.dot(a,b))
print(np.dot(a,b).shape)
```

```python
zeros = np.zeros((3,4))
print("Zeros:\n",zeros)
ones = np.ones((2,3))
print("ones:\n",ones)
identity = np.eye(3)
print("Identity:\n",identity)
```

```python
flat = matrix.flatten()
print("Flatten matrix:",flat)
reshaped = flat.reshape(2,3)
print("Reshaped:\n",reshaped)
```

```python
print("Original:\n",matrix)
transposed = matrix.T
print("Transposed:\n",transposed)
```

```python
v1 = np.array([1,2])
v2 = np.array([3,1])

dot_product = np.dot(v1,v2)
print("v1 * v2:",dot_product)

v3 = np.array([2,4])
print("v1 * v3:",np.dot(v1,v3))
```
