```python
import numpy as np
house1 = np.array([2000, 3, 250000])
house2 = np.array([2200, 4, 300000])
print("House 1:",house1)
print("House 2:",house2)
```

```python
similarity = np.dot(house1, house2)
print("House similarity score:",similarity)

norm_h1 = house1 / np.linalg.norm(house1)
norm_h2 = house2 / np.linalg.norm(house2)
cosine_sim = np.dot(norm_h1, norm_h2)
print("Cosine similarity:",cosine_sim)
```

```python
distance = np.linalg.norm(house1 - house2)
print("Eulidean distance:",distance)

manhattan = np.sum(np.abs(house1 - house2))
print("Manhattan distance:",manhattan)
```

```python
print("Raw ranges:")
print("sqft range:",house1[0],"to",house2[0])
print("price range:",house1[2],"to",house2[2])

def normalize_feature(feature1, feature2):
    min_val = min(feature1, feature2)
    max_val = max(feature1, feature2)
    return (feature1 - min_val) / (max_val - min_val)

norm_h1 = np.array([normalize_feature(house1[0], house2[0]), 
                    normalize_feature(house1[1], house2[1]),
                    normalize_feature(house1[2], house2[2])])
print("Normalized house1:",norm_h1)
```
