```python
import pandas as pd

data = {"Category":["A","B","A","B","A","C"],
        "Sales":[100,200,150,250,200,300],
        "Profit":[10,20,15,25,20,30]
        }

df = pd.DataFrame(data)
df
```

```python
avg_sales = df.groupby("Category")["Sales"].mean()
avg_sales
```

```python
total = df.groupby("Category")[["Sales","Profit"]].sum()
total
```

```python
summary = df.groupby("Category").agg({
    "Sales": "max",
    "Profit": "sum"
})
summary
```
