Generated random Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="darkgrid")

np.random.seed(42)
sq_ft = np.random.normal(1500, 400, 100)
age = np.random.randint(1,50,100)
price = (sq_ft *200) - (age * 1000) + np.random.normal(0, 25000, 100)

df = pd.DataFrame({"Square_Footage": sq_ft, "House_age": age, "House_Price": price})
df.head()
```

Histogram

```python
plt.figure(figsize=(8,5))
sns.histplot(df["House_Price"], bins=15, kde=True, color="blue")
plt.title("Distribution of House Price")
plt.xlabel("Price in USD")
plt.ylabel("Number of Houses")
plt.show()
```

ScatterPlot

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="Square_Footage", y="House_Price", color="green", alpha=0.7)
plt.title("Impact of square Footage on Price")
plt.show()
```

Correlation Heatmap

```python
plt.figure(figsize=(6,4))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```
